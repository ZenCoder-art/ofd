```python
import matplotlib.pyplot as plt
from scipy.stats import binom
import numpy as np
import scienceplots
import seaborn as sns

plt.style.use(["science", "no-latex", "notebook"])

p = 0.6  
n_values = range(1, 101, 2)  
pmv_values = []

for n in n_values:
    k = (n + 1) // 2
    pmv = binom.sf(k - 1, n, p) 
    pmv_values.append(pmv)

fig, ax = plt.subplots(figsize=(8, 6))
sns.lineplot(x=n_values, y=pmv_values, marker='o', color='royalblue', label=f'p = {p}', ax=ax)
plt.xlabel("Number of Voters")
plt.ylabel("Majority Voting Accuracy")
plt.ylim(0.5, 1.05)
ax.tick_params(
    axis="both",
    which="both",
    bottom=False,
    top=False,
    left=False,
    right=False,
    labelbottom=True,
    labelleft=True,
)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scienceplots

plt.style.use(["science", "notebook", "no-latex"])

fake_jobs = {
    'hour': 12206,
    'flexible': 12093,
    'require': 10766,
    'knowledge': 10545,
    'degree': 10322,
    'earn': 10262,
    'week': 10244,
    'establish': 10212,
    'hire': 10208,
    'contact': 10145
}

real_jobs = {
    'work': 48397,
    'experience': 40344,
    'team': 36502,
    'service': 27752,
    'company': 27606,
    'customer': 25977,
    'business': 23167,
    'client': 20178,
    'product': 19252
}

df_fake = pd.DataFrame.from_dict(fake_jobs, orient='index', columns=['count']).reset_index()
df_fake.columns = ['keyword', 'count']
df_fake['type'] = 'Fake Jobs'

df_real = pd.DataFrame.from_dict(real_jobs, orient='index', columns=['count']).reset_index()
df_real.columns = ['keyword', 'count']
df_real['type'] = 'Real Jobs'

df = pd.concat([df_fake, df_real])

fig, axes = plt.subplots(1, 2, figsize=(16, 8))

fake_sorted = df_fake.sort_values('count', ascending=True)
axes[0].barh(fake_sorted['keyword'], fake_sorted['count'], color='#5975A4')
axes[0].set_title('Top 10 Fake Job Keywords', pad=10, fontsize=28)
axes[0].tick_params(
    axis="both",
    which="both",
    bottom=False,
    top=False,
    left=False,
    right=False,
    labelbottom=True,
    labelleft=True,
    labelsize=25
)
real_sorted = df_real.sort_values('count', ascending=True)
axes[1].barh(real_sorted['keyword'], real_sorted['count'], color='#CC8963')
axes[1].set_title('Top 10 Real Job Keywords', pad=10, fontsize=28)
axes[1].tick_params(
    axis="both",
    which="both",
    bottom=False,
    top=False,
    left=False,
    right=False,
    labelbottom=True,
    labelleft=True,
    labelsize=25
)
# for ax in axes:
#     for bar in ax.patches:
#         ax.text(
#             bar.get_width() + 0.5, 
#             bar.get_y() + bar.get_height()/2,
#             f'{int(bar.get_width()):,}',
#             va='center',
#             ha='left',
#             fontsize=15
#         )

plt.tight_layout()
plt.show()
```
